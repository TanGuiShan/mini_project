# Overview of Genre & Related Sections

## For all anime

### Genre

```python
genre_list = [col for col in animelist.columns if 'Genre_' in col]
related_list = [col for col in animelist.columns if 'Related_' in col]
```

```python
def get_count(list, df):   
    list_count=df[list].apply(pd.Series.value_counts).transpose()
    list_count['percentage'] = (list_count[1]/(list_count[0]+list_count[1])*100).round(2)
    list_count= list_count.sort_values(by='percentage', ascending=False)
    list_count=list_count.reset_index()
    return list_count
    
genre_count = get_count(genre_list, animelist)
genre_count
```

```python
# Create the bar plot
sb.catplot(x='percentage', y='index', kind='bar', data=genre_count, height=15, aspect=1.2)
plt.xlabel('Percentage')
plt.ylabel('Genre')
plt.title('Percentage of Each Genre')
plt.show()
```

### Related

```python
def get_related_count(related_list, list):
    related_count = get_count(related_list, list)
    related_count_extra = related_count.drop(columns=['index',0,1,'percentage'])
    related_count['>1'] = related_count[related_count_extra.columns].sum(axis=1)
    related_count = related_count.drop(columns=related_count_extra.columns)
    
    related_count['percentage']=((related_count[1]+related_count['>1'])/(related_count['>1']+related_count[0]+related_count[1])*100).round(2)
    related_count= related_count.sort_values(by='percentage', ascending=False)
    
    related_count['percentage >1']=(related_count['>1']/(related_count['>1']+related_count[0]+related_count[1])*100).round(2)
    return related_count

related_count = get_related_count(related_list, animelist)
related_count
```

```python
# Create the bar plot
sb.catplot(x='percentage', y='index', kind='bar', data=related_count, height=10, aspect=1.2)
plt.xlabel('Percentage')
plt.ylabel('Related')
plt.title('Percentage of Each Related Type')
plt.show()
```

# Looking at Format

## Season

```python
season_df = animelist[['start_season', 'success']]
season_df = season_df.loc[season_df['start_season']!='0']
result = pd.DataFrame(season_df.groupby('start_season')['success'].value_counts(normalize=True).mul(100).round(2))
result = result.sort_values(by=['start_season', 'success'], ascending=True)

sorted_result = result.sort_index(level='success', ascending=True)
sorted_result
```

```python
# Plotting
plt.figure(figsize=(10, 10)) 
sorted_result.plot(kind='bar', stacked=True)
plt.xlabel('Season')
plt.ylabel('Percentage of Success')
plt.title('Percentage of Success in Each Season')
plt.legend(title='Success Level')
plt.xticks(rotation=45)
plt.show()
```

Fall and Summer have a slight edge over Spring & Winter. Note that the distribution is skewed and its only worth comparing between seasons in each success bracket

## Studio

```python
studio_df = animelist[['studio', 'success']]
studio_df['studio'].value_counts()
```

```python
result = pd.DataFrame(studio_df.groupby('studio')['success'].value_counts(normalize=True).mul(100).round(2))
result
```

There are too many studios (unlike seasons), so we will just take top 20 studios with most success <2 to look at generally successful studios

```python


# Filter the DataFrame to include only success levels less than 2
df_success_less_than_2 = studio_df[studio_df['success'] < 2]

# Calculate the percentage of success levels less than 2 for each season
success_less_than_2_percentage = df_success_less_than_2.groupby('studio').size() / studio_df.groupby('studio').size() * 100

# Sort the seasons based on the percentage of success levels less than 2
sorted_seasons = pd.DataFrame(success_less_than_2_percentage.sort_values(ascending=False)).head(20).reset_index()
sorted_seasons = sorted_seasons.rename(columns={0: "percent"})

# Plotting
plt.figure(figsize=(10, 6))
plt.barh(sorted_seasons['studio'], sorted_seasons['percent'], color='skyblue')
plt.xlabel('Percentage of Successful Outcomes')
plt.ylabel('Studios')
plt.title('Percentage of Generally Succesful (Success <2) Outcomes by Studio')
plt.gca().invert_yaxis()  # Invert y-axis to have the highest percentage at the top
plt.show()
```

As expected, well known studios lik Kyoto Animation, Bones, UFOTABLE have good success rates. Though this may not be representative for small studios due to the large number of shows under it.
